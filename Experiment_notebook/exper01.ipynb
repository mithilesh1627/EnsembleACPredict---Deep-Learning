##### Creating the Custom Dataset and DataLoader in pytorch

```python
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
class ACPDataset(Dataset):
    def __init__(self,X_data,y_data):
        if hasattr(X_data, 'values'):
            self.X = torch.tensor(X_data.values, dtype=torch.float32)
            self.y = torch.tensor(y_data.values, dtype=torch.float32).view(-1, 1)
        else:
            self.X = torch.tensor(X_data, dtype=torch.float32)
            self.y = torch.tensor(y_data, dtype=torch.float32).view(-1, 1)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
train_dataset = ACPDataset(X_train,y_train)
test_dataset = ACPDataset(X_test,y_test)
```

```python
import torch.nn as nn
import torch.nn.functional as F

class ACP_CNN_BiLSTM(nn.Module):
    def __init__(self, input_dim=930, conv_filters=64, lstm_units=64):
        super().__init__()
        
        
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=conv_filters, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm1d(conv_filters)
        self.pool = nn.MaxPool1d(conv_filters)

        self.lstm = nn.LSTM(input_size=conv_filters, hidden_size=lstm_units, 
                            num_layers=1, batch_first=True, bidirectional=True)
        
        
        self.fc = nn.Sequential(
            nn.Linear(lstm_units * 2, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = x.unsqueeze(1) 
        
        # CNN path
        x = F.relu(self.bn1(self.conv1(x)))
        x = self.pool(x)
        x = x.transpose(1, 2) 
        
        # BiLSTM path
        lstm_out, (h_n, c_n) = self.lstm(x)
        
        out = torch.cat((h_n[-2,:,:], h_n[-1,:,:]), dim=1)
        
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        return self.sigmoid(self.output(x))
```

```python
import mlflow
import mlflow.pytorch
from sklearn.metrics import accuracy_score, matthews_corrcoef, roc_auc_score

def train_and_log_model(X_train, y_train, X_val, y_val, hp):
    mlflow.set_experiment("ACP_Accuracy_Boost")
    
    with mlflow.start_run(run_name=f"filters_{hp['filters']}_lr_{hp['lr']}"):
        # 1. Initialize Model, DataLoaders, and Optimizer
        # (Assuming you use the ACPFeatureDataset/DataLoader classes we built earlier)
        model = ACP_CNN_BiLSTM(hp)
        optimizer = torch.optim.Adam(model.parameters(), lr=hp['lr'])
        criterion = nn.BCELoss()
        
        mlflow.log_params(hp) # Track hyperparameters

        for epoch in range(50): # Use 50-100 epochs for deeper training
            model.train()
            # ... (Training Loop Logic) ...
            
            # 2. Evaluation
            model.eval()
            with torch.no_grad():
                preds = model(torch.tensor(X_val, dtype=torch.float32)).numpy()
                y_pred_bin = (preds > 0.5).astype(int)
                
                # Calculate the 3 metrics required for your paper
                acc = accuracy_score(y_val, y_pred_bin)
                mcc = matthews_corrcoef(y_val, y_pred_bin)
                auc = roc_auc_score(y_val, preds)

                # 3. Log to MLflow
                mlflow.log_metric("accuracy", acc, step=epoch)
                mlflow.log_metric("mcc", mcc, step=epoch)
                mlflow.log_metric("auroc", auc, step=epoch)

        mlflow.pytorch.log_model(model, "model")
        print(f"Final Acc: {acc:.4f} | MCC: {mcc:.4f}")
```
